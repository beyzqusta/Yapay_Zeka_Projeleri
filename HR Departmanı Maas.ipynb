### Polynominal Linear Regression

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model  import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
df=pd.read_csv("polynomial.csv",sep=";")
```

```python
polynomial_regression=PolynomialFeatures(degree=2)
#bir adet polynomial regression nesnesi oluşturmak için Polynomial Features fonks çağırıyoruz.
#polinom derecesini belirtiyoruz
```

```python
plt.scatter(df["deneyim"],df["maas"])
plt.xlabel("Deneyim(yıl)")
plt.ylabel("maas")
plt.savefig("1.png",dpi=300)
plt.show()
```

```python
# Görüldüğü gibi doğrusal bir yapıda dağılmıyor.
#Eğer biz bu veri setine linear regression uygularsak uygun olmayan bir tahmin elde edilir.
```

```python
reg=LinearRegression()
reg.fit(df[["deneyim"]],df["maas"])

plt.xlabel("deneyim yılı")
plt.ylabel("Maas")
plt.scatter(df["deneyim"],df["maas"])

xekseni=df["deneyim"]
yekseni=reg.predict(df[["deneyim"]])
plt.plot(xekseni,yekseni,color="green",label="linear regression")
plt.legend()
plt.show()
```

```python
polynomial_regression=PolynomialFeatures(degree=4)
```

```python
x_polynomial=polynomial_regression.fit_transform(df[["deneyim"]])
```

```python
reg=LinearRegression()
reg.fit(x_polynomial,df["maas"])
```

```python
y_head=reg.predict(x_polynomial)
plt.plot(df["deneyim"],y_head,color="red",label="polynomial regression")
plt.legend()

plt.scatter(df["deneyim"],df["maas"])
plt.show()
```

```python
# Görüldüğü gibi kesinlikle uygun bir regression seçtik.
```

```python
#degree değiştirerek tekrar deneyelim
```

```python
polynomial_regression=PolynomialFeatures(degree=10)
x_polynomial=polynomial_regression.fit_transform(df[["deneyim"]])
reg=LinearRegression()
reg.fit(x_polynomial,df["maas"])
y_head=reg.predict(x_polynomial)
plt.plot(df["deneyim"],y_head,color="red",label="polynomial regression")
plt.legend()

plt.scatter(df["deneyim"],df["maas"])
plt.show()
```
